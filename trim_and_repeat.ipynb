```python
def trim_and_repeat(text, offset=0, repetitions=1):
    trimmed = text[offset:]
    result = trimmed * repetitions
    return result

# Тест
print(trim_and_repeat("hello", 1, 1))  # Выводится: ello
print(trim_and_repeat("test", 0, 2))   # Выводится: testtest
print(trim_and_repeat("python", 4, 2))  # Выводится: onon
```
